```python
#Checking missing values 

missing_values = df.isnull().sum()
print(missing_values)
```

```python
print((missing_values/ len(df)) * 100)
```

There are two columns with more 20% missing values and two columns with less than 5% missing values. Imputing all of them as these could be critical for analysis.  

- Ways to impute:

1. Imputing 'name' and 'host_name' as "Unknown"
2. Imputing 'last_review' with Timestamp placeholder as "2000-01-01"
3. Imputing 'reviews_per_month' with "0"


```python
#Imputing 'name' and 'host_name' as "Unknown"

df['name'].fillna('Unknown', inplace=True)
df['host_name'].fillna('Unknown', inplace=True)
```

```python
#Imputing 'last_review' with Timestamp placeholder as "2000-01-01"

df['last_review'].fillna(pd.Timestamp('2000-01-01'), inplace=True)
```

```python
#Imputing 'reviews_per_month' with "0"

df['reviews_per_month'].fillna(0, inplace=True)
```

```python
#Checking missing values after imputing

missing_values = df.isnull().sum()
print(missing_values)
```

All the missing values have been imputed. 

```python
#Cleaning categorical columns (standardizing text formats)

df['room_type'] = df['room_type'].str.strip().str.lower()
```

```python
#Ensuring correct data types

df['last_review'] = pd.to_datetime(df['last_review'], dayfirst=True)
```
